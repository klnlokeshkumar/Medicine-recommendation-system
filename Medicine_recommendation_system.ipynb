### Label Encoding and model selection by testing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
x = training.drop('prognosis', axis = 1)
y = training['prognosis']
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
np.unique(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=20)
```

```python
x_train.head()
```

```python
y_train[0: 5]
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
models = {
    "SVC": SVC(kernel='linear'),
    "RandomForestClassifier": RandomForestClassifier(n_estimators=100, random_state=42),
    "GradientBoostingClassifier": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "KNeighborsClassifier": KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB": MultinomialNB()
}
```

```python
for model_name, model in models.items():
    print(model_name, ":", model)
```

```python
import time

for model_name, model in models.items():
    start_time = time.time()
    model.fit(x_train, y_train)
    y_predictions = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_predictions)
    # Confusion matrix construction
    cn = confusion_matrix(y_test, y_predictions)
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"{model_name} accuracy : {accuracy}")
    # print(f"{model_name} confusion matrix : ")
    # print(np.array2string(cn, separator=", "))
    print(f"{model_name} execution time: {elapsed_time:.4f} seconds")
    print("\n\n")
```

### on observation Support vector classifier is choosen 
since it's execution takes less time and more accuracy

```python
svc = SVC(kernel='linear')
svc.fit(x_train, y_train)
y_pred_svc = svc.predict(x_test)
acc_score = accuracy_score(y_test, y_pred_svc)
print(acc_score)
```
